```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/tested.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
titanic_data = df.drop(['PassengerId', 'Cabin', 'Ticket'], axis=1)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='Set1')
plt.title('Survival by Pclass')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=titanic_data, palette='Set2')
plt.title('Survival by Sex')
plt.show()
```

```python
sns.histplot(x='Age', hue='Survived', data=titanic_data, kde=True, palette='Set3', bins=30)
plt.title('Distribution of Age by Survival')
plt.show()
```
